```python
import pandas as pd
from collections import defaultdict
```

```python

# Pfad zur Tabelle
table_path = '../data/RBP2GO/Table_HS_RBP.txt'

# Tabelle mit pandas parsen
df = pd.read_csv(table_path, delimiter='\t')
```

```python

# Ausgabe der Tabelle
print(df.shape)
print(df.columns)
print(df.iloc[1:10, :])
# print(df[['Entry_Name', 'String_ID', 'String_PPI', 'String_PPI_Scores']].head(2))
```

```python
def gather_PPIs(df_, RBPs: dict, PPINs: list):
    current_protein = df_['Uniprot_ID']
    interactors = str(df_['String_PPI']).split('; ')
    PPIN_scores = str(df_['String_PPI_Scores']).split('; ')
    RBPs[current_protein] = {
        'Protein_Name': df_['Protein_Name'],
        'gene_name': df_['Gene_Name'],
        'Alias_Names': df_['Alias_Names'],
        'List_Homologs' : df_['List_Homologs'],
        'List_RBP_Homologs': df_['List_RBP_Homologs']
    }
    
    for i, interactor in enumerate(interactors):
        if interactor > current_protein:
            PPINs.append((current_protein, interactor, PPIN_scores[i]))

        

RBPs = defaultdict(dict)
RBP_interactions = list()
df.apply(gather_PPIs, axis = 1, result_type='expand', RBPs=RBPs, PPINs=RBP_interactions)
```

```python
pd.DataFrame(RBP_interactions).to_csv('../data/RBP2GO/RBP_interactions.csv', index=False, sep='\t', header=False)
```
